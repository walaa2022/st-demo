```python
def extr_profit(x):
    casual=x['casual']*20
    registered=x['registered']*5
    profit=casual+registered
    profit=profit-(0.14*profit)
    maint=1500/(365*24)
    total_profit=profit-maint
    return total_profit

```

```python
df['Profit']=df[['casual','registered']].apply(extr_profit,axis=1)
df
```
